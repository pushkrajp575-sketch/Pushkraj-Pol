```python
import matplotlib.pyplot as plt

labels = ["IT", "HR", "CEO", "MANAGER"]
sizes = [40, 50, 25, 75]

# Top colors
colors = ["#4CAF50", "#2196F3", "#FFC107", "#F44336"]

# Shadow/base colors (darker)
shadow_colors = ["#2E7D32", "#1565C0", "#FF8F00", "#B71C1C"]

plt.figure(figsize=(8, 8))

# Bottom layer (creates 3D thickness)
plt.pie(
    sizes,
    colors=shadow_colors,
    radius=1,
    center=(0, -0.08),      # Move slightly down
    startangle=45
)

# Top layer
plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    radius=1,
    center=(0, 0),
    autopct="%1.1f%%",
    startangle=45,
    shadow=True,
    wedgeprops={"edgecolor": "white", "linewidth": 2}
)

plt.title("Company Report (3D Floating Pie Chart)", fontsize=16)
plt.axis("equal")

plt.show()
```
